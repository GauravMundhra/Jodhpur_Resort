```python
def get_latlng(loc):
    lat_lng_coords = None
    while(lat_lng_coords is None):
        g = geocoder.arcgis('{}, Jodhpur, India'.format(loc))
        lat_lng_coords = g.latlng
    return lat_lng_coords
```

```python
coords = [ get_latlng(loc) for loc in jodh_loc_df["location"].tolist() ]
coords
```

```python
df_coords = pd.DataFrame(coords, columns=['Latitude', 'Longitude'])
jodh_loc_df['Latitude']=df_coords['Latitude']
jodh_loc_df['Longitude']=df_coords['Longitude']
jodh_loc_df
```

```python
venues_df = pd.DataFrame(venues)
venues_df.columns = ['location', 'Latitude', 'Longitude', 'VenueName', 'VenueLatitude', 'VenueLongitude', 'VenueCategory']
print(venues_df.shape)
venues_df.head()
```

```python
venues_df.groupby(["location"]).count()
venues_df['VenueCategory'].unique()[:50]
jodh_onehot = pd.get_dummies(venues_df[['VenueCategory']], prefix="", prefix_sep="")
```

```python
jodh_onehot['location'] = venues_df['location'] 
fixed_columns = [jodh_onehot.columns[-1]] + list(jodh_onehot.columns[:-1])
jodh_onehot = jodh_onehot[fixed_columns]
print(jodh_onehot.shape)
jodh_onehot.head()
```

```python
jodh_grouped = jodh_onehot.groupby(["location"]).mean().reset_index()
print(jodh_grouped.shape)
jodh_grouped
jodh_hotels = jodh_grouped[["location","Hotel","Resort"]]
jodh_hotels.head()
```

```python
from sklearn.cluster import KMeans
clusters = 3
jodh_clustering = jodh_hotels.drop(["location"], 1)

kmeans = KMeans(n_clusters=clusters, random_state=0).fit(jodh_clustering)

kmeans.labels_[0:10]
jodh_merged = jodh_hotels.copy()

jodh_merged["Cluster Labels"] = kmeans.labels_
jodh_merged.head()
```

```python
jodh_merged = jodh_merged.join(jodh_loc_df.set_index("location"), on="location")
jodh_merged.head()
```

```python
jodh_merged.sort_values(["Cluster Labels"], inplace=True)
jodh_merged
```

```python
import numpy as np
import matplotlib.cm as cm
import matplotlib.colors as colors
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)

x = np.arange(clusters)
ys = [i+x+(i*x)**2 for i in range(clusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

markers_colors = []
for lat, lon, poi, cluster in zip(jodh_merged['Latitude'], jodh_merged['Longitude'], jodh_merged['location'], jodh_merged['Cluster Labels']):
    label = folium.Popup(str(poi) + ' - Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
